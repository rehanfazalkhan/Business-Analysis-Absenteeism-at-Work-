```python
import numpy as np
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator,TransformerMixin


#the custom scaler class 
class CustomScaler(BaseEstimator,TransformerMixin):
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler=StandardScaler(copy,with_mean,with_std)
        self.columns=columns
        self.mean_=None
        self.var_=None
        
    def fit(self,X,y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_=np.array(np.mean(X[self.columns]))
        self.var_=np.arrray(np.var(X[self.columns]))
        
    def transform(self,X,y=None,copy=None):
        init_col_order=X.columns
        X_scaled=pd.DataFrame(self.scaler.transform(X[self.columns]),columns=self.columns)
        x_not_scaled=X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled],axis=1)[init_col_order]
        
   


 class absenteeism_model():
    
    def __init__(self,model_file,scaler_file):
        #read the model & scaler where it is save
        with open('model','rb') as model_file,with open('scaler_file','rb') as scaler_file:
            self.reg=pickle.load(model_file)
            self.scaler=pickle.load(scaler_file)
            self.data=None
            
    #take a .csv file and preprocessit
    def load_and_clean_data(self,data_file):
        
        #import the dataset
        df=pd.read_csv(data_file,delimiter=',')
        #store the data into new varibale for later use
        self.df_with_predictions=df.copy()
        #drop the ID column 
        df=df.drop(['ID'],axis=1)
        
        #create the seprate dataframe,contaning dummy variable for ALL available reason
        reason_columns=pd.get_dummies(df['Reason for absence'],drop_first=True)
        
        #split columns into 4 columns
        reason_type_1=reason_columns.loc[:,:14].max(axis=1)
        reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
        reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
        reason_type_4=reason_columns.loc[:,22:].max(axis=1)
        
        #to avoid multi-colinearity drop 'Reason for absence' columns
        df=df.drop(['Reason for absence'],axis=1)
        
        #concatenate the columns
        df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
        
        #assing the name of the columns
        columns=['Month of absence', 'Day of the week', 'Seasons',
       'Transportation expense', 'Distance from Residence to Work',
       'Service time', 'Age', 'Work load Average/day ', 'Hit target',
       'Disciplinary failure', 'Education', 'Son', 'Social drinker',
       'Social smoker', 'Pet', 'Weight', 'Height', 'Body mass index',
       'Absenteeism time in hours', 'reason_1', 'reason_2','reason_3','reason_4']
        
        df.columns=columns
        
        #re-order the columns name in df
        columns_re_order=columns=['reason_1', 'reason_2','reason_3','reason_4',
                                  'Month of absence', 'Day of the week', 'Seasons',
       'Transportation expense', 'Distance from Residence to Work',
       'Service time', 'Age', 'Work load Average/day ', 'Hit target',
       'Disciplinary failure', 'Education', 'Son', 'Social drinker',
       'Social smoker', 'Pet', 'Weight', 'Height', 'Body mass index',
       'Absenteeism time in hours']
        
        df=df['columns_re_order']
        
        #map education varibale 
        df['Education']=df['Education'].map({1:0,2:1,3:1,4:1})
        
        #we don't need this columns
        df=df.drop(['Distance from Residence to Work','Day of the week','Work load Average/day'],axis=1)
        #drop the orginal absent time
        df=df.drop(['Absenteeism time in hours'],axis=1)
        
        self.preprocesed_data=df.copy()
        
        #we need this line so we coluld use it in next function
        self.data=self.scaler.transform(df)
        
    # a function which output the probability of a data point to be 1
    def predicted_probability(self):
        if (self.data is not None):
            pred=self.reg.predict_proba(self.data)[:,1]
            return pred
        
    #a function will have output 0  or 1
    def predicted_output_category(self):
        if (self.data is not None):
            pred_outputs=self.reg.predict(self.data)
            return pred_outputs
        
    def prediction_output(self):
        if(self.data is not None):
            self.preprocesed_data['prediction']=self.reg.predict(self.data)
            self.preprocesed_data['probability']=self.reg.predict_proba(self.data)[:,1]
        
        
```
